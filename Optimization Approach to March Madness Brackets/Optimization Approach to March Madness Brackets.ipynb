```python
#load in data
import pandas as pd
data = pd.read_csv('March Madness.csv')
```

```python
#initialize data frame of probabilities
import numpy as np
probs = pd.DataFrame(np.zeros((64, 6)), index = data.index, columns = ['Round of 64', 'Round of 32', 'Sweet 16', 'Elite 8', 'Final 4', 'National Championship'])
```

## Part b: Formulation of IP

### Decision Variables:
$x_{i, j}$ = 1, if picking team i to win in round j; 0, otherwise

### Data:
$p_{i, j}$ = probability of team i winning in round j

### Objective Function:
max $\sum_{i=1}^{64} x_{i, 1}p_{i, 1} + 2x_{i, 2}p_{i, 2} + 4x_{i, 3}p_{i, 3} + 8x_{i, 4}p_{i, 4} + 16x_{i, 5}p_{i, 5} + 32x_{i, 6}p_{i, 6}$

### Constraints:
#### Variable Constraints:
$x_{i, j}$ binary $\forall$ i = 1, ..., 64; j = 1, ..., 6

#### Restriction on picking team in following round if eliminated:
$x_{i, j} \leq x_{i, j-1}$ $\forall$ i = 1, ..., 64; j = 2, ..., 6

#### Must pick exactly one winner for each game in round 1:
$x_{i, 1} + x_{i+1, 1} = 1$ $\forall$ i = 1, 3, 5, ..., 61, 63

#### Must pick exactly one winner for each game in round 2:
$x_{i, 2} + x_{i+1, 2} + x_{i+2, 2} + x_{i+3, 2} = 1$ $\forall$ i = 1, 5, 9, ..., 61

#### Must pick exactly one winner for each game in round 3:
$\sum_{n = 0}^7 x_{i+n, 3} = 1$ $\forall$ i = 1, 9, 17, ..., 57

#### Must pick exactly one winner for each game in round 4:
$\sum_{n = 0}^{15} x_{i+n, 4} = 1$ $\forall$ i = 1, 17, 33, 49

#### Must pick exactly one winner for each game in round 5:
$\sum_{n = 0}^{31} x_{i+n, 5} = 1$ $\forall$ i = 1, 33

#### Must pick exactly one winner for round 6:
$\sum_{i = 1}^64 x_{i, 6} = 1$

## Part c: Solving Model in PuLP

```python
#import pulp
from pulp import *
```

```python
#Set up problem
prob = LpProblem("MarchMadness", LpMaximize)
```

```python
#Define decision variables
var1 = LpVariable.dicts("x1", teams, 0, 1, cat = 'binary')
var2 = LpVariable.dicts("x2", teams, 0, 1, cat = 'binary')
var3 = LpVariable.dicts("x3", teams, 0, 1, cat = 'binary')
var4 = LpVariable.dicts("x4", teams, 0, 1, cat = 'binary')
var5 = LpVariable.dicts("x5", teams, 0, 1, cat = 'binary')
var6 = LpVariable.dicts("x6", teams, 0, 1, cat = 'binary')
```

```python
#Elimination constraints
for team in teams:
    prob += var6[team] <= var5[team]
    prob += var5[team] <= var4[team]
    prob += var4[team] <= var3[team]
    prob += var3[team] <= var2[team]
    prob += var2[team] <= var1[team]
```

```python
#Game winner constraints
#Round 1 winners:
for i in range(0, 64, 2):
    prob += var1[teams[i]] + var1[teams[i+1]] == 1
prob += sum([var1[team] for team in teams]) == 32
#Round 2 winners:
for i in range(0, 64, 4):
    prob += var2[teams[i]] + var2[teams[i+1]] + var2[teams[i+2]] + var2[teams[i+3]] == 1
prob += sum([var2[team] for team in teams]) == 16
#Round 3 winners:
for i in range(0, 64, 8):
    prob += var3[teams[i]] + var3[teams[i+1]] + var3[teams[i+2]] + var3[teams[i+3]] + var3[teams[i+4]] + var3[teams[i+5]] + var3[teams[i+6]] + var3[teams[i+7]] == 1
prob += sum([var3[team] for team in teams]) == 8
#Round 4 winners:
prob += sum([var4[teams[i]] for i in range(16)]) == 1
prob += sum([var4[teams[i]] for i in range(16,32)]) == 1
prob += sum([var4[teams[i]] for i in range(32,48)]) == 1
prob += sum([var4[teams[i]] for i in range(48,64)]) == 1
prob += sum([var4[team] for team in teams]) == 4
#Round 5 winners:
prob += sum([var5[teams[i]] for i in range(32)]) == 1
prob += sum([var5[teams[i]] for i in range(32,64)]) == 1
prob += sum([var5[team] for team in teams]) == 2
#Round 6 winners:
prob += sum([var6[teams[i]] for i in range(64)]) == 1   
```

```python
#Objective Function
prob += 1*(sum([var1[team]*probs['Round of 64'][team] for team in teams])) + 2*(sum([var2[team]*probs['Round of 32'][team] for team in teams])) + 4*(sum([var3[team]*probs['Sweet 16'][team] for team in teams])) + 8*(sum([var4[team]*probs['Elite 8'][team] for team in teams])) + 16*(sum([var5[team]*probs['Final 4'][team] for team in teams])) + 32*(sum([var6[team]*probs['National Championship'][team] for team in teams]))
```

```python
#Set up Gurobi Solver
solver = getSolver('GUROBI')
```

```python
#Solve problem
prob.solve(solver)
```

```python
#Set up problem
prob = LpProblem("MarchMadness", LpMaximize)

#Define decision variables
var1 = LpVariable.dicts("x1", teams, 0, 1, cat = 'binary')
var2 = LpVariable.dicts("x2", teams, 0, 1, cat = 'binary')
var3 = LpVariable.dicts("x3", teams, 0, 1, cat = 'binary')
var4 = LpVariable.dicts("x4", teams, 0, 1, cat = 'binary')
var5 = LpVariable.dicts("x5", teams, 0, 1, cat = 'binary')
var6 = LpVariable.dicts("x6", teams, 0, 1, cat = 'binary')

#Elimination constraints
for team in teams:
    prob += var6[team] <= var5[team]
    prob += var5[team] <= var4[team]
    prob += var4[team] <= var3[team]
    prob += var3[team] <= var2[team]
    prob += var2[team] <= var1[team]

#Game winner constraints
#Round 1 winners:
for i in range(0, 64, 2):
    prob += var1[teams[i]] + var1[teams[i+1]] == 1
prob += sum([var1[team] for team in teams]) == 32
#Round 2 winners:
for i in range(0, 64, 4):
    prob += var2[teams[i]] + var2[teams[i+1]] + var2[teams[i+2]] + var2[teams[i+3]] == 1
prob += sum([var2[team] for team in teams]) == 16
#Round 3 winners:
for i in range(0, 64, 8):
    prob += var3[teams[i]] + var3[teams[i+1]] + var3[teams[i+2]] + var3[teams[i+3]] + var3[teams[i+4]] + var3[teams[i+5]] + var3[teams[i+6]] + var3[teams[i+7]] == 1
prob += sum([var3[team] for team in teams]) == 8
#Round 4 winners:
prob += sum([var4[teams[i]] for i in range(16)]) == 1
prob += sum([var4[teams[i]] for i in range(16,32)]) == 1
prob += sum([var4[teams[i]] for i in range(32,48)]) == 1
prob += sum([var4[teams[i]] for i in range(48,64)]) == 1
prob += sum([var4[team] for team in teams]) == 4
#Round 5 winners:
prob += sum([var5[teams[i]] for i in range(32)]) == 1
prob += sum([var5[teams[i]] for i in range(32,64)]) == 1
prob += sum([var5[team] for team in teams]) == 2
#Round 6 winners:
prob += sum([var6[teams[i]] for i in range(64)]) == 1   

#part h constraint
prob += var3[teams[0]] + var3[teams[14]] + var3[teams[16]] + var3[teams[30]] + var3[teams[32]] + var3[teams[46]] + var3[teams[48]] + var3[teams[62]] <= 5

#Objective Function
prob += 1*(sum([var1[team]*probs['Round of 64'][team] for team in teams])) + 2*(sum([var2[team]*probs['Round of 32'][team] for team in teams])) + 4*(sum([var3[team]*probs['Sweet 16'][team] for team in teams])) + 8*(sum([var4[team]*probs['Elite 8'][team] for team in teams])) + 16*(sum([var5[team]*probs['Final 4'][team] for team in teams])) + 32*(sum([var6[team]*probs['National Championship'][team] for team in teams]))
```

```python
#Set up problem
prob = LpProblem("MarchMadness", LpMaximize)

#Define decision variables
var1 = LpVariable.dicts("x1", teams, 0, 1, cat = 'binary')
var2 = LpVariable.dicts("x2", teams, 0, 1, cat = 'binary')
var3 = LpVariable.dicts("x3", teams, 0, 1, cat = 'binary')
var4 = LpVariable.dicts("x4", teams, 0, 1, cat = 'binary')
var5 = LpVariable.dicts("x5", teams, 0, 1, cat = 'binary')
var6 = LpVariable.dicts("x6", teams, 0, 1, cat = 'binary')

#Elimination constraints
for team in teams:
    prob += var6[team] <= var5[team]
    prob += var5[team] <= var4[team]
    prob += var4[team] <= var3[team]
    prob += var3[team] <= var2[team]
    prob += var2[team] <= var1[team]

#Game winner constraints
#Round 1 winners:
for i in range(0, 64, 2):
    prob += var1[teams[i]] + var1[teams[i+1]] == 1
prob += sum([var1[team] for team in teams]) == 32
#Round 2 winners:
for i in range(0, 64, 4):
    prob += var2[teams[i]] + var2[teams[i+1]] + var2[teams[i+2]] + var2[teams[i+3]] == 1
prob += sum([var2[team] for team in teams]) == 16
#Round 3 winners:
for i in range(0, 64, 8):
    prob += var3[teams[i]] + var3[teams[i+1]] + var3[teams[i+2]] + var3[teams[i+3]] + var3[teams[i+4]] + var3[teams[i+5]] + var3[teams[i+6]] + var3[teams[i+7]] == 1
prob += sum([var3[team] for team in teams]) == 8
#Round 4 winners:
prob += sum([var4[teams[i]] for i in range(16)]) == 1
prob += sum([var4[teams[i]] for i in range(16,32)]) == 1
prob += sum([var4[teams[i]] for i in range(32,48)]) == 1
prob += sum([var4[teams[i]] for i in range(48,64)]) == 1
prob += sum([var4[team] for team in teams]) == 4
#Round 5 winners:
prob += sum([var5[teams[i]] for i in range(32)]) == 1
prob += sum([var5[teams[i]] for i in range(32,64)]) == 1
prob += sum([var5[team] for team in teams]) == 2
#Round 6 winners:
prob += sum([var6[teams[i]] for i in range(64)]) == 1   

#part h constraint second variation
prob += var3[teams[10]] + var3[teams[26]] + var3[teams[42]] + var3[teams[58]] >= 1
prob += var2[teams[6]] + var2[teams[22]] + var2[teams[38]] + var2[teams[54]] <= 2
prob += var2[teams[4]] + var2[teams[8]] + var2[teams[12]] + var2[teams[20]] + var2[teams[24]] + var2[teams[28]] + var2[teams[36]] + var2[teams[40]] + var2[teams[44]] + var2[teams[52]] + var2[teams[56]] + var2[teams[60]] >= 1

#Objective Function
prob += 1*(sum([var1[team]*probs['Round of 64'][team] for team in teams])) + 2*(sum([var2[team]*probs['Round of 32'][team] for team in teams])) + 4*(sum([var3[team]*probs['Sweet 16'][team] for team in teams])) + 8*(sum([var4[team]*probs['Elite 8'][team] for team in teams])) + 16*(sum([var5[team]*probs['Final 4'][team] for team in teams])) + 32*(sum([var6[team]*probs['National Championship'][team] for team in teams]))
```
